# Pengaturan Format Pada Plot (parameter ketiga)

Pengaturan parameter ketiga -> gaya formatting pada MATLAB
- Nilai default -> 'b-'
- Pemanggil fungsi axis() menyertakan sebuah list `[xmin, xmax, ymin, ymax]`

```python
plt.plot([1,2,3,4],[1,4,9,16],'r+') 
#--> warna merah (r), warna biru(b), warna hijau (g)
#--> o,*,-,+,etc menunjukan bentuk grafik
plt.axis([0,6,0,20]) #[xmin, xmax, ymin, ymax]
plt.show()
```

# MULTI PLOT

```python
t = np.arange(0.,5.,0.2) #(nilai min, nilai max, interval)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
#bentuk lain multi plot
plt.plot(t, t, 'r--',
         t, t**2, 'bs',
         t, t**3, 'g^')
plt.show()
```

# PLOTTING DENGAN KEYWORDS

```python
data = { 'a': np.arange(50),
       'c':np.random.randint(0, 50, 50),
       'd':np.random.randn(50)}
data['b'] = data['a']+10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
#dimensi ditampilkan menurut warna(c) dan ukuran(s)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
