# Simple Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

# Pengaturan Grid dan Color

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

- Mengubah warna pada saat penggunaan plt.bar() -> `color = warna, alpha = transparansi`
- Mengubah grid dengan `plt.grid()` -> `linewidth = ketebalan grid, axis = arah grid sesuai sumbu`

```python
plt.bar(kategori, data,
       color='red',
       alpha=0.25)

plt.grid(linestyle='--',
        linewidth=2,
        axis='y',
        alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.xlabel('Contoh Bar Plot')

plt.show()
```

# Horizontal Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.barh(kategori, data)
#plt.barh(kategori[::-1],data[::-1])

plt.grid(linestyle='--',
        linewidth=1,
        axis='x',
        alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

Membalikan Urutan Bar Plot

```python
plt.barh(kategori[::-1],data[::-1])

plt.grid(linestyle='--',
        linewidth=1,
        axis='x',
        alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

# Stacked Bar Plot

Stacked Bar Plot : `Untuk melakukan plotting 2 data set berbeda yang memiliki kesamaan dalam pembagian kategori`

2 Buah Kumpulan Nilai dengan Kategori sama

```python
data1 = [20, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A','B','C','D','E']
```

```python
plt.bar(kategori, data1, label='Data1')
plt.bar(kategori, data2, label='Data2', bottom=data1) #bottom : spesifikasi penempatan

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

Perbedaan penempatan bottom
- `Tanpa bottom maka akan ada penimpahan plotting`

```python
plt.bar(kategori, data1, label='Data1', bottom=data2) #bottom : spesifikasi penempatan
plt.bar(kategori, data2, label='Data2')

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
plt.bar(kategori, data1, label='Data1')
plt.bar(kategori, data2, label='Data2')
plt.bar(kategori, data2, label='Data2', bottom=data1) #bottom : spesifikasi penempatan

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

# Grouped Bar Plot

```python
data1 = [20, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A','B','C','D','E']

x = np.arange(len(kategori))
x
```

Harus definisikan `lebar (width) -> 0.0 - 1.0`

```python
width = 0.35
plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.grid(linestyle='--',
        linewidth=1,
        axis='y',
        alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
```

Akan ditampilkan tepat berdampingan satu dengan lainnya
