# Kasus Twin Axes 1

`linspace`(nilai terkecil, nilai terbesar, jumlah dataset)

```python
x = np.linspace(1, 10, 25)
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp') 
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x), 'ro-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```

Penjelasan plotting (menggunakan OO Style) :
- terdapat 1 sumbu x dan 2 sumbu y
- sumbu y ke 1 (kiri) berkorelasi dengan nilai exponensial
- sumbu y ke 2 (kanan) berkorelasi dengan nilai logaritma
