# Import Module

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

# Simple Line Plot

```python
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
s = np.sin(2*np.pi*x)
s
```

```python
# Object Oriented Style
fig, ax = plt.subplots()
ax.plot(x, s)
ax.set(xlabel='nilai x',
      ylabel='nilai sin(x)',
      title='Visualisasi nilai sin')
ax.grid() #menambahkan grid pada plotting - optional
plt.show()
```

^^^ Object Oriented Style

```python
# Pyplot Style
plt.plot(x,s,label="Visualisasi Nilai Sin")
plt.xlabel('nilai x')
plt.ylabel('nilai sin(x)')
plt.grid()
plt.show()
```

^^^ Pyplot Style

# Multiple Subplots

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)
# (nilai minimum<inclusive>, nilai maximum<inclusive>, data poin yang akan di generate)

y1 = np.cos(2*np.pi*x1)
y2 = np.cos(2*np.pi*x2)
```

```python
x1
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai cos(x1)')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)')

ax2.set_label('Nilai x')

plt.show()
```

Multiple Subplots dengan Object Oriented Style

```python
plt.subplot(2 ,1, 1)
#parameter ketiga (...,...,x) untuk mengakses
plt.plot(x1,y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x-1)$')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

Multiple Subplots dengan Pyplot Style
